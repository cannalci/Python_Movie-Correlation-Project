```python
"""
Data is retrieved from https://www.kaggle.com/danielgrijalvas/movies on 18.12.2021

Check for Correlation between Gross Revenue and Company, Budget, Votes

Data Cleaning, Data Manipulation, Data Analysis, Data Visualization

Python, Pandas, Numpy, Seaborn, Matplotlib, Jupyter
"""


# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

import csv

df = pd.read_csv(r'C:\Users\Acer\Downloads\movies.csv\movies.csv')
```

```python
# Look at the data
df.head()
```

```python
# Check any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# data types for columns
df.dtypes
```

```python
#delete null values and check missing values again

df = df.dropna(how='any',axis=0) 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#Change Data Types
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
#Extract Correct Release Year and Country from Release Column

df['string_released'] = df['released'].astype(str)
df[['split_one', 'split_two']] = df['string_released'].str.split(',', 1, expand=True)
df['yearcorrect'] = df['split_two'].astype(str).str[1:5]
df['countrycorrect'] = df['split_two'].astype(str).str[7:-1]
df
```

```python
# Drop Any Duplicates

df.drop_duplicates()

```

```python
#Sort Values by Gross Revenue and Show All Raws in The Table

pd.set_option('display.max_rows', None)
df = df.sort_values(by=['gross'], inplace=False, ascending=False)

df.head()
```

```python
#Scatter Plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
#  Plot Budget vs Gross Using Seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "orange"}, line_kws={"color": "blue"})
```

```python
# Correlation
df.corr(method='pearson') #pearson, kendall, spearman
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True, fmt='.1g')
plt.title('Correlation Matrix for All Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
